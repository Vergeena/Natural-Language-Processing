```python
def sentimentalAnalysis(dataset,columnName):    
    #downloading vader_lexicon for  the process
    import matplotlib.pyplot as plt
    import seaborn as sns
    "Import neccessary packages"
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    import nltk
    "Deleting null values"
    dataset.dropna(inplace=True)
    "Checking for the comment"
    #sia.polarity_scores(data[columnName][63])
    "Creating respective columns"
    dataset['scores']=dataset[columnName].apply(lambda commentText:sia.polarity_scores(commentText))
    dataset['compound']=dataset['scores'].apply(lambda score_dict:score_dict['compound'])
    dataset['positive']=dataset['scores'].apply(lambda score_dict:score_dict['pos'])
    dataset['neutral']=dataset['scores'].apply(lambda score_dict:score_dict['neu'])
    dataset['negative']=dataset['scores'].apply(lambda score_dict:score_dict['neg'])
    "Creating final pos or neg using compound score"
    dataset['comp_score']=dataset['compound'].apply(lambda c:'pos' if c>=0 else 'neg')
    plt.clf()
    #comp=sns.countplot(x='comp_score',hue='Name',data=data,palette='magma')
    #comp.figure.savefig("date_charts.png")
    "Checking how many pos or neg"
    posneg=pd.DataFrame(dataset['comp_score'].value_counts())
    return posneg,dataset
```

```python
pos,dataset_senti=sentimentalAnalysis(dataset,columnName='Chat')
```

# Talkative & Less talkative

```python
# Step 1: Read the WhatsApp chat data
df = pd.read_csv('whatsapp_data.csv')

# Display the first few rows to understand the structure
print(df.head())

# Step 2: Define labels based on criteria (example: message length)
def classify_talkativeness(message):
    # Example criteria: classify as talkative if message length is long
    if len(message.split()) >= 5:
        return 'talkative'
    else:
        return 'less talkative'

# Apply classification function to each row
df['talkativeness_prediction'] = df['Chat'].apply(classify_talkativeness)

# Display the predicted results
print("Predicted Talkativeness:")
print(df[['Date', 'Time', 'Name', 'Chat', 'talkativeness_prediction']])

# Optionally, you can save the results to a new CSV file
#df.to_csv('whatsapp_data_with_talkativeness_predictions.csv', index=False)
```

# Most active time ,date ,hours

```python
# Step 1: Read the WhatsApp chat data
df = pd.read_csv('whatsapp_data.csv')

# Display the first few rows to understand the structure
print(df.head())

# Step 2: Convert 'Date' and 'Time' to datetime format
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Step 3: Extract day of the week and hour of the day
df['DayOfWeek'] = df['DateTime'].dt.day_name()
df['HourOfDay'] = df['DateTime'].dt.hour

# Step 4: Calculate activity metrics
most_active_day = df['DayOfWeek'].mode()[0]  # Mode gives the most frequent value
most_active_time = df['HourOfDay'].mode()[0]  # Mode gives the most frequent value

# Display results
print(f"Most Active Day: {most_active_day}")
print(f"Most Active Time: {most_active_time}:00 - {most_active_time + 1}:00")
```

# Media Count

```python
# Step 1: Read the WhatsApp chat data
df = pd.read_csv('whatsapp_data.csv')

# Display the first few rows to understand the structure
print(df.head())

# Step 2: Define function to identify media messages
def media_message(chat_text):
    return '<Media omitted>' in str(chat_text)  # Check if message contains media indicator

# Step 3: Mark messages containing media
df['Media SR count'] = df['Chat'].apply(media_message)

# Step 4: Group by 'Name' and count media messages
media_counts = df[df['Media SR count']].groupby('Name').size().reset_index(name='MediaCount')

# Display results
print("Media Counts:")
print(media_counts)
```
